## Na amostragem sistemática, nós dividimos a **população pela amostra**
 > A **amostragem sistemática** é um tipo de amostragem **probabilística**, onde se faz uma **seleção aleatória do primeiro elemento** para a amostra e logo se **selecionam os itens subsequentes utilizando intervalos fixos** ou sistemáticos até chegar ao tamanho da amostra desejada.

 > ### Podemos assim dizer, que os métodos de seleção sistemática das unidades de amostra ***não se baseiam*** na teoria de amostragem probabilística **pois ela seleciona somente uma unidade de amostra ao acaso**

 Exemplo:
> população -> [1,20,3,40,5,62,7,48,96,105,]   
> quero 2 numeros como amostra
> Vamos dividir o tamanho do grupo pela quantidade de amostra.
> 10 / 2 = 5  
> O primeiro é randômico -> Exemplo: foi sorteado o 40
>
> O primeiro numero da amostra é [40]
> o segundo é o 40 + 5 posições = 96
>
> Amostra = [40,96]

```python
def amostragem_sistematica(dataset,amostras):
    intervalo = len(dataset) // amostras
    
    #random.seed(1) # fixa o valor randomico
    inicial = random.randint(0 , intervalo )
    indices_randomicos = np.arange(inicial, len(dataset), step= intervalo)
    return dataset.iloc[indices_randomicos]
```
