## Na amostragem sistemtica, ns dividimos a **populao pela amostra**
 > A **amostragem sistemtica**  um tipo de amostragem **probabilstica**, onde se faz uma **seleo aleatria do primeiro elemento** para a amostra e logo se **selecionam os itens subsequentes utilizando intervalos fixos** ou sistemticos at chegar ao tamanho da amostra desejada.

 > ### Podemos assim dizer, que os mtodos de seleo sistemtica das unidades de amostra ***no se baseiam*** na teoria de amostragem probabilstica **pois ela seleciona somente uma unidade de amostra ao acaso**

 Exemplo:
> populao -> [1,20,3,40,5,62,7,48,96,105,]   
> quero 2 numeros como amostra
> Vamos dividir o tamanho do grupo pela quantidade de amostra.
> 10 / 2 = 5  
> O primeiro  randmico -> Exemplo: foi sorteado o 40
>
> O primeiro numero da amostra  [40]
> o segundo  o 40 + 5 posies = 96
>
> Amostra = [40,96]

```python
def amostragem_sistematica(dataset,amostras):
    intervalo = len(dataset) // amostras
    
    #random.seed(1) # fixa o valor randomico
    inicial = random.randint(0 , intervalo )
    indices_randomicos = np.arange(inicial, len(dataset), step= intervalo)
    return dataset.iloc[indices_randomicos]
```
